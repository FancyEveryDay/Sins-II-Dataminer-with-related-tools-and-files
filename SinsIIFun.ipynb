```python
shipList = getSinsData("",".unit")
weaponList = getSinsData("",".weapon")
weaponDict = createWeaponDict(weaponList)
```

```python
# Calculate toughest ship by supply, no abilities or tech. lvl 1
toughest = [("",0)] # (name,eHP per supply)

for i in shipList:
    unitName = i[0].split("\\")[-1]
    ship = i[1]

    if "ship_roles" in ship:

        try: unitSupply = ship["build"]["supply_cost"]
        except: unitSupply = 0

        try: unitDurability = ship["health"]["durability"]
        except: unitDurability = 0
            
        try: unitArmorStrength = ship["health"]["levels"][0]["armor_strength"]   
        except: unitArmorStrength = 0

        # try: unitHP = ship["health"]["levels"][9]["max_hull_points"]
        # except: 
        try: unitHP = ship["health"]["levels"][0]["max_hull_points"]
        except: unitHP = 0   

        # try: unitAP = ship["health"]["levels"][9]["max_armor_points"] 
        # except:
        try: unitAP = ship["health"]["levels"][0]["max_armor_points"] 
        except: unitAP = 0

        # try: unitSP = ship["health"]["levels"][9]["max_shield_points"]     
        # except: 
        try: unitSP = ship["health"]["levels"][0]["max_shield_points"]  
        except: unitSP = 0

        generalEhp = (unitHP + unitSP) * ((100 + unitDurability) / (100))

        armorEhp = (unitAP) * ((100 + unitDurability) /(100)) * ((100 + unitArmorStrength) / (100))

        totalEhp = generalEhp + armorEhp

        try:
            ehpPerSupply = totalEhp / unitSupply
        except: ehpPerSupply = 0

        
        toughest.append((unitName, ehpPerSupply))

toughest.sort(key = lambda i: i[1], reverse = True)

print(f"the {toughest[0][0]} is the toughest ship at {toughest[0][1]}eHP per supply.")
print(f"{"Rank" : <4} - {"Unit" : <34} {"eHP / supply"}")
for i, unit in enumerate(toughest): print(f"{i : <4} - {unit[0] : <34} {str(int(round(unit[1],0))) : >4}")
```

```python
# Calculate killiest ship by supply, no abilities or tech. Lvl 1
# def avgDamage(dps, penitration, targetDurability):

killiest = [("",0)] # (name,eHP per supply)

for i in shipList:
    unitName = i[0].split("\\")[-1]
    ship = i[1]

    level = 3
    targetDurability = 500


    if "ship_roles" in ship:
        totalDPS = 0

        try: unitSupply = ship["build"]["supply_cost"]
        except: unitSupply = 0

        try: unitWeaponsBlock = ship["weapons"]["weapons"]  
        except: unitWeaponsBlock = False #or maybe make it false and add a conditional

        try: 
            carrier = ship["carrier"]["base_max_squadron_capacity"] 
            try: carrier += int(ship["levels"]["levels"][level-1]["unit_modifiers"]["additive_values"]["max_squadron_capacity"])
            except: pass
            
        except: carrier = False

        try:
            weaponDamage = ship["levels"]["levels"][level-1]["weapon_modifiers"]["scalar_values"]["damage"]
            weaponFireRate = ship["levels"]["levels"][level-1]["weapon_modifiers"]["scalar_values"]["cooldown_duration"]
        except: 
            weaponDamage = 0
            weaponFireRate = 0

        weaponCollection = {}

        if unitWeaponsBlock is not False:
            weaponCollection = {}
            for weapon in unitWeaponsBlock:
                weapon = weapon['weapon']
                if "planet" not in weapon:
                    if weapon in weaponCollection:
                        weaponCollection[weapon] += 1
                    else:
                        weaponCollection[weapon] = 1

        if carrier is not False:
            if "vasari" == unitName.split("_")[0]:
                weaponCollection["vasari_bomber_strikecraft_heavy_phase_missile"] = 3 * carrier
            
            if "trader" == unitName.split("_")[0]:
                weaponCollection["trader_interceptor_strikecraft_light_autocannon"] = 6 * carrier

            if "advent" == unitName.split("_")[0]:
                weaponCollection["advent_interceptor_strikecraft_light_laser"] = 8 * carrier

        for weapon, count in weaponCollection.items():
            weaponName : str = weaponDict[weapon]["name"].split(".")[-1]
            weaponName = weaponName.replace("_", " ")

            weaponDPS : float = weaponDict[weapon]['dps']
            # weaponDPS = round(weaponDPS, 1)
            
            try: weaponPenetration = weaponDict[weapon]["penetration"]
            except: weaponPenetration = 0

            if targetDurability is False:
                targetList = [0, 50, 150, 200, 300, 500, 600, 750, 1000]
                resultList = []

                for i in targetList:
                    if (i - weaponPenetration) > 0:
                        modifiedDurability = i - weaponPenetration
                    else: modifiedDurability = 0

                    resultList.append(weaponDPS * (100 / (100 + modifiedDurability)))

                weaponDPS = sum(resultList) / len(resultList)

            else:
                if (targetDurability - weaponPenetration) > 0:
                    modifiedDurability = targetDurability - weaponPenetration
                else: modifiedDurability = 0

                # weaponRange = weaponDict[weapon]["range"]
                
                weaponDPS = weaponDPS * (100 / (100 + modifiedDurability))

            if "strikecraft" not in weapon:
                weaponDPS = weaponDPS * ((1 + weaponDamage) / 1) / (1 / (1 - weaponFireRate))

            totalDPS += weaponDPS * count

        try:
            dpsPerSupply = totalDPS / unitSupply
        except: dpsPerSupply = 0

        
        killiest.append((unitName, dpsPerSupply))

killiest.sort(key = lambda i: i[1], reverse = True)

print(f"the {killiest[0][0]} is the killiest ship vs {targetDurability} durability at {round(killiest[0][1],2)}DPS per supply.")
print(f"{"Rank" : <4} - {"Unit" : <34} {"DPS / supply"}")
for i, unit in enumerate(killiest): print(f"{i+1 : <4} - {unit[0] : <34} {str(round(unit[1],2)) : <4}")
```
