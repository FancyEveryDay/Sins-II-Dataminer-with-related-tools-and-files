```python
from SinsIIStatsThing import getSinsData, createWeaponDict
import json, pprint, glob
```

```python
with open(r'Patch 42.05 Localized Text\en.localized_text', 'r') as file:
    LOCALIZED_TEXT = json.load(file)

playerGlob = glob.glob(r'Patch 42.05 Entities\**.player')
```

```python
playerDict = {}

for fileName in playerGlob:
    fileName : str
    with open(fileName, 'r') as file:
        faction = json.load(file)

    try:
        race = LOCALIZED_TEXT.get("player_race." + faction['race'] ,faction['race'].capitalize())
        name = "player_faction_name." + fileName.split("\\")[-1].split('.')[0]
        raceFaction = LOCALIZED_TEXT[name]

    except KeyError:
        continue

    playerDict[raceFaction] = {'research' : faction["research"]["research_subjects"],
                               'planet_components' : faction["planet_components"],
                        'name' : name,
                        'race' : race}
```

```python
subjectGlob = glob.glob(r'Patch 42.05 Entities\**.research_subject')

subjectDict = {}

for i in subjectGlob:
    with open(i, 'r') as file:
        subject = json.load(file)

    name = i.split("\\")[-1].replace(".research_subject", "")
    subjectDict[name] = subject.get('name')

pprint.pprint(subjectDict)
```

```python
componentGlob = glob.glob(r'Patch 42.05 Entities\**.unit_item')

componentDict = {}

for i in componentGlob:
    with open(i, 'r') as file:
        component = json.load(file)

    name = i.split("\\")[-1].replace(".unit_item", "")
    componentDict[name] = component.get('name')

pprint.pprint(componentDict)
```

```python
wiki_player = {}
for faction, factionDict in playerDict.items():
    if wiki_player.get(factionDict.get('race')) == None:
        wiki_player[factionDict.get('race')] = {'research_subjects' : [], 'planet_components' : []}
        print(factionDict.get('race'))

    wiki_player[factionDict.get('race')][faction] = factionDict
    print('\t' + faction)
    
```

```python

## Research Subjects
for faction, itemList in playerDict.items():
    race = itemList['race']
    uniqueResearchList = itemList['unique_research_subjects'] = []
    print(faction)

    for item in itemList.get('research'):
        item_is_copy = False

        for f, i in playerDict.items():
            if (f == faction) or (itemList.get('name').find("dlc_") < 0 and i.get('name').find("dlc_") > -1):
                continue

            for j in i.get('research'):
                if item == j:
                    item_is_copy = True
                    break

            if item_is_copy == True:
                try: 
                    string = f'{race} {LOCALIZED_TEXT[subjectDict[item]]}'
                except:
                    print(f"Oopsie with {item} string on finding item is copy")

                if string not in wiki_player[race]['research_subjects']:
                    wiki_player[race]['research_subjects'].append(string)

                break

        if item_is_copy == False:
            try: 
                string = f'{race} {LOCALIZED_TEXT[subjectDict[item]]}'
                uniqueResearchList.append(string)
                print("\t" + string)
            except: 
                string = item
                uniqueResearchList.append(string)
                print("\t" + string)

for faction, itemList in playerDict.items():
    itemList.pop('research')
    itemList.get('unique_research_subjects').sort()

for race, itemList in wiki_player.items():
    itemList.get('research_subjects').sort()
```

```python
for faction, itemList in playerDict.items():
    race = itemList['race']
    uniquePlanetItemList = itemList['unique_planet_components'] = []
    print(faction)

    for item in itemList.get("planet_components"):
        item_is_copy = False

        for f, i in playerDict.items():
            if (f == faction) or (itemList.get('name').find("dlc_") < 0 and i.get('name').find("dlc_") > -1):
                continue

            for j in i.get('planet_components'):
                if item == j:
                    item_is_copy = True
                    break

            if item_is_copy == True:
                try: 
                    string = f'{race} {LOCALIZED_TEXT[componentDict[item]]}'
                except:
                    print(f"Oopsie with {item} string on finding item is copy")

                if string not in wiki_player[race]['planet_components']:
                    wiki_player[race]['planet_components'].append(string)

                break

        if item_is_copy == False:
            try: 
                string = f'{race} {LOCALIZED_TEXT[componentDict[item]]}'
                uniquePlanetItemList.append(string)
                print("\t" + string)
            except: 
                string = item
                uniquePlanetItemList.append(string)
                print("\t" + string)

for faction, itemList in playerDict.items():
    itemList.pop('planet_components')
    itemList.get('unique_planet_components').sort()

for race, itemList in wiki_player.items():
    itemList.get('planet_components').sort()
```

```python
pprint.pprint(wiki_player)

with open('WikiPlayer.json', 'w') as file:
    json.dump(wiki_player, file, indent = 1)
```
