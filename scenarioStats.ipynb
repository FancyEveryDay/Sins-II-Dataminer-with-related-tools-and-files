```python
from SinsIIStatsThing import getSinsData, createWeaponDict, FormatUnitEntries, SINS_DIRECTORY, LAST_PATCH_NUM
import json, glob, shutil, pprint
```

```python
scenarios = glob.glob(SINS_DIRECTORY + "\\scenarios\\public**.scenario")
```

```python
print(f"{'Map' : ^30} : Gas Giants | City | Geomagnetic | Super Rich | Weighted | Unweighted")
for map in scenarios:
    # if "random" in map:
    #     continue

    with open(map, 'r', encoding="utf-8", errors="ignore") as f:
        mapname = (map.split("\\")[-1]
                    .replace(".scenario", "")
                    .replace("public_", "")
                    .replace("_", " ")
                    .capitalize())
        
        gasGiants = 0
        city = 0
        magnetic = 0
        superRich = 0
        weighted = 0
        unweighted = 0
    
        for line in f:
            try:
                if "gas_giant" in line:
                    gasGiants += 1
                if "city_" in line:
                    city += 1
                if "magnetic" in line:
                    magnetic += 1
                if "super_rich" in line:
                    superRich += 1  
                if "weighted" in line:
                    weighted += 1
                if "unweighted" in line:
                    unweighted += 1
            except:
                continue
        
        if gasGiants + city + magnetic + superRich + weighted + unweighted > 0:
            print(f"{mapname : >30} : {gasGiants : ^10} | {city : ^4} | {magnetic : ^11} | {superRich : ^10} | {weighted : ^8} | {unweighted : ^10}")
```
